```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd

audio_file, sr = librosa.load("audio/oboe_c6.wav")
print(audio_file.shape)
```

```python
plt.figure(figsize=(15,5))
librosa.display.waveshow(audio_file, sr=sr)
```

```python
audio_spec = librosa.stft(audio_file)
audio_spec_db = librosa.amplitude_to_db(abs(audio_spec))
plt.figure(figsize=(15,5))
librosa.display.specshow(audio_spec_db, sr=sr, x_axis="time", y_axis="hz")
```

```python
ipd.Audio(audio_file, rate=sr)
```
